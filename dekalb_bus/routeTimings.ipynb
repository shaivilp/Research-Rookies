```python
df = pl.read_parquet("./data/2024-09-entries.parquet")
df = df.with_columns(pl.col("routeID").replace_strict(route_mapping))

df
```

```python
# Load stops data
file = open("./data/stops.json", "r")
stopsData = json.load(file)

stops = pl.DataFrame(stopsData['get_stops'])
stops
```

```python
#Load patterns data
pattern_mapping = {
    3: "2L",
    4: "2R",
    37: "3",
    17: "10",
    18: "11",
    23: "12",
    12: "16",
    13: "17",
    14: "18",
    33: "19",
    46: "21",
    45: "21 Tripper",
}

# Load patterns json
file = open("./data/patterns.json", "r")
patternsData = json.load(file)

patterns = pl.DataFrame(patternsData['get_patterns'])
patterns = patterns.with_columns(pl.col("id").replace_strict(pattern_mapping, default="None"))
patterns
```

```python
stops.rename({"id": "nextStopID"}).select(["nextStopID", "name"]).unique().sort("nextStopID")
```

```python
#Left join stops to get stop names for nextStopID and lastStopID
df = df.join(stops.rename({"id": "nextStopID"}).select(["nextStopID", "name"]).unique(), on="nextStopID", how="left").rename({"name": "nextStopName"})
df = df.join(stops.rename({"id": "lastStopID"}).select(["lastStopID", "name"]).unique(), on="lastStopID", how="left").rename({"name": "lastStopName"})

df
```

```python
#Looking to figure out why we had such big huge max timeDiffs between stops
sub = df.filter(
    (pl.col("routeID") == "2L") &
    (pl.col("equipmentID") == "1002") &
    (pl.col("receiveTime") >= datetime.datetime(2024, 9, 3, 15, 54))
)

sub = sub.with_columns(
    (pl.col("nextStopID").diff().over("equipmentID") != 0).alias("stopChanged")
)

sub
```

```python
#Calculate the time between each stop in the 2L Route

#Select all columns with that routeId and inService
subset = df.filter(
    (pl.col("routeID") == "2L") &
    (pl.col("inService"))
)

#Add stopChanged column
subset = subset.with_columns(
    (
        (pl.col("nextStopID") == pl.col("lastStopID").shift(-1)).over(
            "equipmentID", order_by="receiveTime"
        )
    ).alias("stopChanged")
)

subset = subset.filter(pl.col("stopChanged")).with_columns(
    (pl.col("nextStopID").shift(1) == pl.col("lastStopID"))
    .over("equipmentID", order_by="receiveTime")
    .alias("nextToLast")
)

#Calculate time diffs between all stops
allDiffs = subset.with_columns(
    (pl.col("receiveTime").diff().over("equipmentID", order_by="receiveTime")).alias(
        "timeDiff"
    )
).filter(pl.col("nextToLast"))

#Add timeDiff column
subset = subset.with_columns(
    (pl.col("receiveTime").diff().over("equipmentID", order_by="receiveTime")).alias(
        "timeDiff"
    )
).filter(pl.col("nextToLast"))

#Filter out any timedeltas above 30 minutes
# subset = subset.filter((pl.col("timeDiff") < datetime.timedelta(minutes=30)))
# allDiffs = subset

#Aggregate table with the mean, median, min, and max timeDiffs
subset = subset.group_by("lastStopID", "nextStopID").agg(
        pl.col("lastStopName").first(),
        pl.col("nextStopName").first(),
        pl.col("timeDiff").mean().alias("meanTimeDiff"),
        pl.col("timeDiff").median().alias("medianTimeDiff"),
        pl.col("timeDiff").min().alias("minTimeDiff"),
        pl.col("timeDiff").max().alias("maxTimeDiff"),
)

subset
```
