```python
import polars as pl
pl.enable_string_cache()
pl.Config().set_tbl_cols(100)
pl.Config().set_tbl_rows(50)
```

```python
%config InteractiveShell.ast_node_interactivity = 'last_expr_or_assign'
```

```python
two_l_data = pl.read_parquet("../data/two_l_mega_df_variance.parquet")

two_l_data = two_l_data.drop_nulls()
```

```python
two_l_data.filter((pl.col("lastStopID") == 820) & (pl.col("nextStopID_actual") == 433)).select(pl.col("eta_seconds").quantile(.95))
```

```python
twol_variance_df = (
    two_l_data
    .group_by(["lastStopID", "nextStopID_actual"])
    .agg([
        pl.col("eta_seconds").var().alias("variance"),
        pl.col("lastStopName").first(),
        pl.col("nextStopName_actual").first(),
        pl.col("eta_seconds").mean().alias("mean_eta"),
        pl.col("eta_seconds").median().alias("median_eta"),
        pl.col("eta_seconds").min().alias("min_eta"),
        pl.col("eta_seconds").max().alias("max_eta"),
        pl.col("distance").first()
    ])
    .sort(["lastStopName", "variance"], descending=True)
)
```

```python
twelve_data = pl.read_parquet("../data/twelve_route_mega_df.parquet")
```

```python
twelve_ariance_df = (
    twelve_data
    .group_by(["lastStopID", "nextStopID_actual"])
    .agg([
        pl.col("eta_seconds").var().alias("variance"),
        pl.col("lastStopName").first(),
        pl.col("nextStopName_actual").first()
    ])
    .sort("variance", descending=True)
)
```
