```python
import polars as pl
import json

pdf = pl.read_parquet('./data/2024-09-preds.parquet')
```

```python
weirdData = pdf.filter(pl.col('schedule').str.contains('-'))
weirdData
```

```python
pdf_dt = pdf.with_columns(
    pl.col('receiveTime').dt.convert_time_zone('America/Chicago'),
    pl.col('schedule').alias('scheduleStr'),
    pl.col('time').alias('timeStr'),
    pl.col('schedule').str.to_time(format='%I:%M%p', strict=False),
    pl.col('time').str.to_time(format='%I:%M%p', strict=False)
)
```

```python
#Load stops data
file = open("./data/stops.json", "r")
stopsData = json.load(file)

stops = pl.DataFrame(stopsData['get_stops'])
stops
```

```python
stops.rename({"id": "nextStopID"}).select(["nextStopID", "name"]).unique().sort("nextStopID")
```

```python
#Left join stops to get stop names for nextStopID and lastStopID
pdf_dt = pdf_dt.join(stops.rename({"id": "stopID"}).select(["stopID", "name"]).unique(), on="stopID", how="left").rename({"name": "stopName"})
```

```python
#Check for scheudle and time that are null or weird
pdf_dt_weird = pdf_dt.filter(
  pl.col('time').is_null()
)

pdf_dt_weird
```

```python
pdf_dt2 = pdf_dt.with_columns(pl.col('receiveTime').dt.combine(pl.col('time')).alias('time'),
                    pl.col('receiveTime').dt.combine(pl.col('schedule')).alias('schedule'))
```

```python
pdf_diff = pdf_dt2.with_columns((pl.col('time') - pl.col('schedule')).alias('scheduleDiff'))
```

```python
pdf_by_status = pdf_diff.group_by('statuscolor').agg(pl.col('scheduleDiff').min().alias('scheduleDiffMin'),
                                     pl.col('scheduleDiff').max().alias('scheduleDiffMax'),
                                     pl.col('scheduleDiff').median().alias('scheduleDiffMedian'))
```

```python
pl.Config.set_tbl_rows(30)
```

```python
import datetime

min_mins = -10
max_mins = 19
colors = []
for statuscolor in pdf_diff['statuscolor'].unique():
    # all diffs
    colors.append(pdf_diff.filter((pl.col('scheduleDiff').dt.total_seconds() > 60 * min_mins) & (pl.col('scheduleDiff').dt.total_seconds() < 60 * max_mins) & (pl.col('statuscolor') == statuscolor))['scheduleDiff'].dt.total_seconds().hist(bins=list(range(min_mins*60,max_mins*60,60))).with_columns(pl.lit(statuscolor).alias('statuscolor')))
    # only diffs where the lag is low
    # colors.append(pdf_diff
    #     .with_columns((pl.col('captureTime').dt.convert_time_zone('America/Chicago') - pl.col('receiveTime')).alias('receiveDiff')).sort('receiveDiff')
    #     .filter(pl.col('receiveDiff') < datetime.timedelta(seconds=20))
    #     .filter((pl.col('scheduleDiff').dt.total_seconds() > 60 * min_mins) & (pl.col('scheduleDiff').dt.total_seconds() < 60 * max_mins) & (pl.col('statuscolor') == statuscolor))['scheduleDiff'].dt.total_seconds().hist(bins=list(range(min_mins*60,max_mins*60,60))).with_columns(pl.lit(statuscolor).alias('statuscolor'))
    # )

pdf_colors = pl.concat(colors)
```

```python
import altair as alt

colors = pdf_diff['statuscolor'].unique().to_list()
color_scale = alt.Color(
    'statuscolor:N',
    scale=alt.Scale(
        domain=colors,
        range=colors
    )
)

pdf_colors.plot.bar(x='category', y='count',color=color_scale)
```
